```python
from bs4 import BeautifulSoup

from splinter import Browser
from splinter.exceptions import ElementDoesNotExist
```

```python
# URL of page to be scraped
url = 'https://www.barnesandnoble.com/bookSearch/isbnInquiry.asp?isbn=1467741876'
```

```python
executable_path = {'executable_path': './chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
browser.visit(url) 
browser
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
soup.prettify
```

```python
bn_price_results = soup.find_all('span', class_='price current-price ml-0')
bn_price_results
```

```python
bn_price = bn_price_results[0].text
bn_price
```

```python
list_price_results = soup.find_all('input', id='listPrice')
list_price_results
```

```python
list_price = list_price_results[0]['value']
list_price
```
