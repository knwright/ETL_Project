```python
from bs4 import BeautifulSoup

from splinter import Browser
from splinter.exceptions import ElementDoesNotExist
```

```python
# URL of page to be scraped
url = 'https://www.amazon.com/dp/1467741876/'
```

```python
executable_path = {'executable_path': './chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
browser.visit(url) 
browser
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
amazon_price_results = soup.find_all('span', class_='a-size-medium a-color-price offer-price a-text-normal')
amazon_price_results
```

```python
amazon_price = amazon_price_results[0].text
amazon_price
```

```python
list_price_results = soup.find_all('span', class_='a-color-secondary a-text-strike')
list_price_results
```

```python
list_price = list_price_results[0].text
list_price
```

```python
# Scrape ISBN
isbn_chunk = soup.find('div', attrs={'class':'content'})
isbn_chunk
```

```python
isbn_li = isbn_chunk.find_all('li')[4].text
# isbn = isbn_li.replace('ISBN-10: ', '')
isbn_li
```
